**Check if the following is prime number or not**

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(" is a prime number.")
else:
    print(" is not a prime number.")
```

**Check if the string is palindrome or not
tahe input from user**

```python
def is_palindrome(input_string):
    cleaned_string = input_string.replace(" ", "").lower()
    return cleaned_string == cleaned_string[::-1]

user_input = input("Enter a string: ")

if is_palindrome(user_input):
    print(" is a palindrome")
else:
    print(" is not a palindrome")
```

Create a empty list and seek 10 elements from user into list and categorize the elements into odd or even. Print the odd and even list seperately

```python
elements = []

for _ in range(10):
    element = int(input("Enter an element= "))
    elements.append(element)

odd_elements = []
even_elements = []

for element in elements:
    if element % 2 == 0:
        even_elements.append(element)
    else:
        odd_elements.append(element)

print("Odd elements:", odd_elements)
print("Even elements:", even_elements)
```
